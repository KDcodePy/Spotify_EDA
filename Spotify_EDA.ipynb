# Spotify Top 10000 Songs YR: 1960-Present 

## Imports 

```python
import pandas as pd
import numpy as np
```

## Load the Data

```python
path = 'top_10000_1960-now.csv'
```

```python
df = pd.read_csv(path)
```

## Data Cleaning

```python
df.info()
```

```python
# Check for missing data
df.isna().sum().sort_values(ascending=False)
```

- Droping `Album Genres`, `Track Preview URL`, `Artist Genres` for excessive missing values  

```python
# Drop columns with missing values > 20% of total dataset
df.drop(columns=['Album Genres','Track Preview URL',
                 'Artist Genres'], inplace=True)
```

```python
# Drop missing values <1% of total dataset
df.dropna(inplace=True)
```

```python
display(df.head(), df.tail())
```

```python
# Check for duplicate
df.duplicated().sum()
```

```python
# Remove duplicated data
df.drop_duplicates(inplace=True)
```

```python
# Drop Irrelevant Columns
df.drop(columns=['Track URI', 'Artist URI(s)','Album URI','Album Artist URI(s)',
                 'Album Image URL','Disc Number','Track Number',
                 'Copyrights','ISRC','Added By','Album Artist Name(s)'],inplace=True)
```

```python
df.head()
```

## Feature Engineering 

```python
df['Album Release Date'].apply(lambda x: len(x)).sort_values()
```

```python
# Check format
print(df['Album Release Date'].apply(lambda x: x.split('-')))
```

```python
# Parse YYYY-MM from Release Date
df['Release_Year'] = df['Album Release Date'].apply(lambda x: x.split('-')[0])
df['Release_Month'] = df['Album Release Date'].apply(lambda x: x.split('-')[1] if len(x)>4 else 'NA' )
```
